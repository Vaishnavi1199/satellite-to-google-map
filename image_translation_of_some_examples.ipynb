```
from numpy import asarray
from numpy import vstack
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from numpy import load
```

```
def load_val_example(image_path,size=(256,512)):
  src_list=[]
  tar_list=[]
  
  image=load_img(f'{image_path}',target_size=size)
  np_image= img_to_array(image)
  sat=(np_image[:,:256]/127.5)-1 
  img=(np_image[:,256:]/127.5)-1
  src_list.append(sat)
  tar_list.append(img)
  return [asarray(src_list),asarray(tar_list)]
```

```
def plot_images(src_img,gen_img,tar_img , j):
  images=vstack((src_img , gen_img , tar_img))
  images=(images+1)/2.0
  titles= ['Satellite_img' , 'Generated_gmap' ,"Expected_gmap"]
  fig=plt.figure()
  for i in range(3):
    plt.subplot(1,3,1+i)
    plt.axis('off')
    plt.imshow(images[i])
    plt.title(titles[i])
  filename=(f"/content/drive/My Drive/sat2gmap/val_subplots/image{j}.png")
  fig.savefig(filename)
```

```
model1=load_model("/content/drive/My Drive/sat2gmap/g_model/plot_190.h5")
```

```
image_path1="/content/drive/My Drive/maps /val/27.jpg"
```

```
src_img,tar_img=load_val_example(image_path1)
```

```
gen_img=model1.predict(src_img)
```

```
plot_images(src_img,gen_img,tar_img,1)
```
